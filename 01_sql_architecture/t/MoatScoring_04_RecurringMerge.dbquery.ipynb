```sql
-- Täglicher Snapshot-Load (Merge Pattern)
-- Führen Sie diese Query täglich aus oder schedulen Sie sie als Job
MERGE INTO thekitchen.t.MoatScoring_Snapshot AS target
USING (
  SELECT
    CURRENT_DATE() AS snapshot_date,
    Company AS company,
    System_Role AS system_role,
    CAST(REGEXP_EXTRACT(System_Role, 'SR-(\\d+)', 1) AS INT) AS system_role_id,
    S1_Substitution_Test AS s1_substitution_test,
    S2_Source_of_Enforcement AS s2_source_of_enforcement,
    S3_Replacement_Time AS s3_replacement_time,
    S4_Capital_KnowHow AS s4_capital_knowhow,
    S5_Demand_Enforcement AS s5_demand_enforcement,
    S6_Moat_Decay_Test AS s6_moat_decay_test,
    Moat_Vector_Trend AS moat_vector_trend,
    Aggregated_Score AS aggregated_score,
    Decay_Adjusted_Score AS decay_adjusted_score,
    Classification AS classification,
    Strategic_Takeaway AS strategic_takeaway
  FROM thekitchen.s.moat_scoring
  WHERE is_current = 1
) AS source
ON target.snapshot_date = source.snapshot_date 
   AND target.company = source.company
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
