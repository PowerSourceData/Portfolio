```sql
--Create View from s.two_year_treasury
CREATE OR REPLACE VIEW t.us_treasury_two_years AS
SELECT
  timestamp as Date,
  round(coalesce(value, 0), 2) as Yield,
  "2y_US" as YieldID
FROM
  two_year_treasury
WHERE
  value IS NOT NULL;

--Create View from s.five_year_treasury
CREATE OR REPLACE VIEW t.us_treasury_five_years AS
SELECT
  timestamp as Date,
  round(coalesce(value, 0), 2) as Yield,
  "5y_US" as YieldID
FROM
  five_year_treasury
WHERE
  value IS NOT NULL;

--Create View from s.ten_year_treasury
CREATE OR REPLACE VIEW t.us_treasury_ten_years AS
SELECT
  timestamp as Date,
  round(coalesce(value, 0), 2) as Yield,
  "10y_US" as YieldID
FROM
  ten_year_treasury
WHERE
  value IS NOT NULL;
```
