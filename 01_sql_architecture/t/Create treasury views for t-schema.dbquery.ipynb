```sql
/*
=============================================================================
PURPOSE: US Treasury Yield Curve Data Transformation
=============================================================================
This script creates standardized views for US Treasury yields across different
maturities (2Y, 5Y, 10Y) from raw source tables in schema 's'.

TARGET SCHEMA: thekitchen.t (transformed data layer)
SOURCE SCHEMA: thekitchen.s (staging layer)

BUSINESS VALUE:
- Provides clean, consistent treasury yield data for financial analysis
- Enables yield curve analysis and interest rate trend monitoring
- Supports portfolio risk assessment and macroeconomic forecasting

DATA QUALITY:
- Filters out NULL values to ensure data integrity
- Rounds yields to 2 decimal places for consistency
- Standardizes column naming across all maturity views
=============================================================================
*/

-- ============================================
-- VIEW 1: 2-Year US Treasury Yields
-- ============================================
-- Source: s.two_year_treasury
-- Purpose: Short-term interest rate indicator, often used for near-term economic outlook
CREATE OR REPLACE VIEW t.us_treasury_two_years AS
SELECT
  timestamp as Date,                          -- Original timestamp from source data
  round(coalesce(value, 0), 2) as Yield,     -- Yield percentage, rounded to 2 decimals
  "2y_US" as YieldID                          -- Unique identifier for 2-year maturity
FROM
  two_year_treasury;                          

-- ============================================
-- VIEW 2: 5-Year US Treasury Yields
-- ============================================
-- Source: s.five_year_treasury
-- Purpose: Mid-term interest rate benchmark, key for mortgage rates and corporate bonds
CREATE OR REPLACE VIEW t.us_treasury_five_years AS
SELECT
  timestamp as Date,                          -- Original timestamp from source data
  round(coalesce(value, 0), 2) as Yield,     -- Yield percentage, rounded to 2 decimals
  "5y_US" as YieldID                          -- Unique identifier for 5-year maturity
FROM
  five_year_treasury
WHERE;                          

-- ============================================
-- VIEW 3: 10-Year US Treasury Yields
-- ============================================
-- Source: s.ten_year_treasury
-- Purpose: Long-term interest rate benchmark, widely used reference for economic health
CREATE OR REPLACE VIEW t.us_treasury_ten_years AS
SELECT
  timestamp as Date,                          -- Original timestamp from source data
  round(coalesce(value, 0), 2) as Yield,     -- Yield percentage, rounded to 2 decimals
  "10y_US" as YieldID                         -- Unique identifier for 10-year maturity
FROM
  ten_year_treasury;                          

/*
=============================================================================
USAGE EXAMPLES:
=============================================================================
-- Query recent 2-year yields:
SELECT * FROM t.us_treasury_two_years ORDER BY Date DESC LIMIT 100;

-- Compare all three maturities (yield curve analysis):
SELECT * FROM t.us_treasury_two_years
UNION ALL
SELECT * FROM t.us_treasury_five_years
UNION ALL
SELECT * FROM t.us_treasury_ten_years
ORDER BY Date DESC, YieldID;
=============================================================================
*/
```
