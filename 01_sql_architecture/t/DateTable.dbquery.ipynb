```sql
-- =================================================================
-- View: thekitchen.t.DateTable
-- Zweck: Zentrale Datumsdimension mit PascalCase-Spaltennamen
-- Quelle: thekitchen.s.datetable
-- Beschreibung: Stellt alle Zeitdimensionen bereit (Jahr, Quartal,
--               Monat, Woche, Tag) mit konsistenter Namenskonvention
--               für BI-Tools und Analysen
-- =================================================================
CREATE OR REPLACE VIEW `thekitchen`.`t`.`DateTable` AS
SELECT
  -- Basisdatum
  `Date` -- Primäres Datumfeld (DATE)
  -- Jahr/Monat Dimensionen
  ,`CY_Month` AS `CyMonth` -- Monatsanfang (TIMESTAMP)
  ,`CY_Year` AS `CyYear` -- Jahresanfang (TIMESTAMP)
  ,`Months_Long` AS `MonthsLong` -- Monat mit vollem Namen
  ,`Months` -- Monat (DATE)
  ,`MonthNumber` -- Monatsnummer (1-12)
  -- Quartalsinformationen
  ,`QuarterYear` -- Quartal + Jahr (z.B. "Q3/2026")
  ,`Quarters` -- Quartal ("Q1"-"Q4")
  ,`QuarterYearSort` -- Sortierfeld INT (YYYYQ, z.B. 20263)
  ,`QuartersSort` -- Quartalsnummer 1-4 (INT)
  -- Offset-Berechnungen (relativ zu heute)
  ,`OffsetYear` -- Jahre seit Referenzdatum
  ,`OffsetMonth` -- Monate seit Referenzdatum
  ,`OffsetDay` -- Tage seit Referenzdatum
  -- Zusätzliche Datumsformate
  ,`DateLong` -- Datum in Langform
  ,`Is_EoM` AS `IsEom` -- End-of-Month Flag (BOOLEAN)
  -- Wochendimensionen
  ,`Calendar_Weeks` AS `CalendarWeeks` -- Kalenderwoche ("CW XX")
  ,`Calendar_Weeks_SORT` AS `CalendarWeeksSort` -- KW als Zahl (1-53)
  ,`WeekdayDate` AS `Weekday` -- Wochentag (DATE)
  ,`WeekdayDateLong` AS `WeekdayLong` -- Wochentag Langform
  ,`WeekDayNum` AS `WeekdayNum` -- Wochentagsnummer (1-7)
  ,`DayOfYear` -- Tag im Jahr (1-366)
  -- Arbeitstag-Klassifikationen
  ,`Workday` -- "Workday" / "Weekend"
  ,`BusinessDay` -- "Business day" / "Rest day"
FROM
  datetable -- Quelltabelle: thekitchen.s.datetable (~22k Zeilen, 323 Jahre)
```
