```sql
-- =============================================================================
-- Treasury Yield Views - Data Architecture Layer
-- =============================================================================
-- Purpose: Create staging/consumption views in schema 's' that reference
--          raw data tables in schema 'av' (Alpha Vantage source)
--
-- Architecture Pattern: Source → Staging Views
--   - Source Layer (av schema): Raw data from Alpha Vantage API
--   - Staging Layer (s schema): Abstraction layer
--
-- Benefits:
--   • Decouples consumers from source schema changes
--   • Enables future transformations without breaking downstream queries
--   • Centralizes access control (grant permissions on views, not tables)
--   • Supports multi-source consolidation (if needed later)
--
-- Schema: thekitchen.s (staging layer)
-- Source: thekitchen.av (raw Alpha Vantage data)
-- =============================================================================


-- -----------------------------------------------------------------------------
-- 2-Year Treasury Yield View
-- -----------------------------------------------------------------------------
-- Short-term interest rate indicator (tracks near-term economic expectations)
-- Use Cases: Yield curve analysis, short-term rate modeling, monetary policy analysis

CREATE OR REPLACE VIEW s.two_year_treasury AS
SELECT
  timestamp,                 -- Trading date
  value AS yield_percent,    -- Yield in percent (e.g., 3.45 = 3.45%)
  CreatedAt,                 -- Audit: Initial load timestamp
  UpdatedAt                  -- Audit: Last update timestamp (NULL if never updated)
-- SELECT CONCAT('-- ', FORMAT_NUMBER(COUNT(*), '#,##0'), ' Rows on ', DATE_FORMAT(CURRENT_DATE(), 'dd.MM.yyyy')) AS RowCount_
FROM
  av.two_year_treasury;


-- -----------------------------------------------------------------------------
-- 5-Year Treasury Yield View
-- -----------------------------------------------------------------------------
-- Medium-term interest rate indicator
-- Use Cases: Yield curve construction, intermediate rate analysis

CREATE OR REPLACE VIEW s.five_year_treasury AS
SELECT
  timestamp,                 -- Trading date
  value AS yield_percent,    -- Yield in percent
  CreatedAt,                 -- Audit: Initial load timestamp
  UpdatedAt                  -- Audit: Last update timestamp
-- SELECT CONCAT('-- ', FORMAT_NUMBER(COUNT(*), '#,##0'), ' Rows on ', DATE_FORMAT(CURRENT_DATE(), 'dd.MM.yyyy')) AS RowCount_
FROM
  av.five_year_treasury;


-- -----------------------------------------------------------------------------
-- 10-Year Treasury Yield View
-- -----------------------------------------------------------------------------
-- Benchmark long-term interest rate (most widely watched Treasury maturity)
-- Use Cases: Risk-free rate for CAPM, mortgage rate proxy, economic benchmark
-- Note: Federal Reserve and financial markets use this as key economic indicator

CREATE OR REPLACE VIEW s.ten_year_treasury AS
SELECT
  timestamp,                 -- Trading date
  value AS yield_percent,    -- Yield in percent
  CreatedAt,                 -- Audit: Initial load timestamp
  UpdatedAt                  -- Audit: Last update timestamp
-- SELECT CONCAT('-- ', FORMAT_NUMBER(COUNT(*), '#,##0'), ' Rows on ', DATE_FORMAT(CURRENT_DATE(), 'dd.MM.yyyy')) AS RowCount_
FROM
  av.ten_year_treasury;


-- =============================================================================
-- Next Steps:
--   1. Grant SELECT permissions to analyst roles: GRANT SELECT ON SCHEMA s TO `analysts`
--   2. Use these views in dashboards and downstream queries
--   3. Build yield curve analysis by joining all three views on timestamp
-- =============================================================================
```
